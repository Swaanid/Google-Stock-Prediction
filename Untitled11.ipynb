## Split the data

```python
train_percentage = 0.8

# Calculate the split index based on the training percentage
split_index = int(len(df1) * train_percentage)

# Split the data into training and testing sets
train_data, test_data = df1.iloc[:split_index], df1.iloc[split_index:]

# Print the shapes of the training and testing sets
print("Training data shape:", train_data.shape)
print("Testing data shape:", test_data.shape)
```

```python
df1['close'][:split_index].plot(figsize=(16,4),legend=True)
df1['close'][split_index:].plot(figsize=(16,4),legend=True, color = 'r')
plt.legend(['Training set ','Test set '], fontsize=12)
plt.title('Google stock price', fontsize=16)
plt.show()
```

## Normalize the data

```python
data = df1[['close']].values

# Normalize the data
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data)
```

## Function to create a time series dataset

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)

look_back = 10  
# Create the time series dataset
X, y = create_dataset(data_normalized, look_back)

```

## Build and Compile the LSTM Model

```python
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Reshape the input data for LSTM (samples, time steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(look_back, 1), return_sequences=True))
model.add(LSTM(units=100, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=100, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=100))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])

model.summary()
```

## Train the model

```python
history=model.fit(X, y, epochs=20, batch_size=32, verbose=2)
```

## predictions on the test set

```python
test_target = test_data['close'].values.reshape(-1, 1)

# Normalize the testing data using the same scaler
test_target_scaled = scaler.transform(test_target)

# Create the time series dataset for testing
X_test, y_test = create_dataset(test_target_scaled, look_back)

# Reshape the input data for LSTM
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Make predictions on the testing set
predictions = model.predict(X_test)
```

```python
predictions = model.predict(X_test)

# Invert predictions to original scale
predictions = scaler.inverse_transform(predictions)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
```

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
loss = history.history['loss']

epochs = range(len(loss))

plt.plot(epochs, loss, 'b', label='Training loss')

plt.title('Training loss', size=15, weight='bold')
plt.legend(loc=0)
plt.figure()

plt.show()
```

## Visualisation of LSTM model

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual Prices', color='blue')
plt.plot(predictions, label='Predicted Prices', color='orange')
plt.title('Actual vs Predicted Prices')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
def plot_predictions(test,predicted):
    plt.plot(test, color='red',label='Real Google Stock Price')
    plt.plot(predicted, color='blue',label='Predicted Google Stock Price')
    plt.title('Google Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Google Stock Price')
    plt.legend()
    plt.show()
    
plot_predictions(test_target, predictions)
```

The LSTM model, with a Root Mean Squared Error (RMSE) of 67, exhibits predictive capabilities in forecasting Google stock prices. The architecture comprises multiple LSTM layers with dropout regularization, enabling the model to capture complex temporal dependencies in the time series data. The sequential structure of the model allows it to learn patterns over various time steps, contributing to its ability to make reasonably accurate predictions. The RMSE metric quantifies the average magnitude of prediction errors, and a lower RMSE value signifies a closer alignment between predicted and actual stock prices.
